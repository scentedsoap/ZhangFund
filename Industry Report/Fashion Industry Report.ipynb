```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import requests
import json
pd.options.display.float_format = '{:.2f}'.format   # suppress scientific notation
```

```python
gdp_file_path = "../data/macro economic data/GDP/GDP.csv"
inflation_file_path = "../data/macro economic data/Inflation/Inflation.csv"
real_gdp_file_path = "../data/macro economic data/Constant GDP/ConstantGDP.csv"
real_gdp_growth_file_path = "../data/macro economic data/Growth Rates/Real GDP Growth Rates.csv"

world_gdp_df = pd.read_csv(gdp_file_path, on_bad_lines="warn")
world_inflation_df = pd.read_csv(inflation_file_path)
real_gdp_df = pd.read_csv(real_gdp_file_path)
real_gdp_growth_rates_df = pd.read_csv(real_gdp_growth_file_path)
```

```python
world_real_gdp_df = real_gdp_df["World"]
```

```python
# displaying top 20 Regions by Real GDP in 2022
display(world_gdp_df.sort_values(by="2022", ascending=False)[["Country Name", "2022"]].head(30))
```

```python
us_gdp = world_gdp_df[world_gdp_df["Country Name"] == "United States"]
us_inflation = world_inflation_df[world_inflation_df["Country Name"] == "United States"]

non_year_columns = ["Country Name", "Country Code", "Indicator Name", "Indicator Code", "Unnamed: 67"]

us_gdp_by_year = us_gdp.melt(id_vars=non_year_columns, var_name="years", value_name="gdp")
us_inflation_by_year = us_inflation.melt(id_vars=non_year_columns, var_name="years", value_name="inflation")

display(us_gdp)
display(us_inflation)

display(us_inflation_by_year)
display(us_gdp_by_year)

us_gdp_by_year["real_gdp"] = us_gdp_by_year["gdp"] / (us_inflation_by_year["inflation"] / 100 + 1)
display(us_gdp_by_year)

us_gdp_by_year.plot(x="years",y="gdp")

```
